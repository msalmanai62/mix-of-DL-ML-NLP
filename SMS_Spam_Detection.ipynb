```python
# importing libraries
import nltk
import re
import pandas as pd
# import numpy as np
# from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
# from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC  # SVM
```

# creating objects

```python

# stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
# cv = CountVectorizer()
tfidfv = TfidfVectorizer()
```

# reading data
download link --->  https://archive.ics.uci.edu/ml/machine-learning-databases/00228/smsspamcollection.zip

```python
data = pd.read_csv("/content/drive/MyDrive/SMSSpamCollection.csv", sep="\t", names=["lable", "message"])
```

```python
data
```

```python
data.describe()
```

```python
nltk.download('stopwords')
stopwords.words("english")
```

# Data Preprocessing
- tokenization
- Punctuation removal
- stopwords removal
- lemmatization

```python
nltk.download('punkt')
nltk.download('wordnet')
```

```python
nltk.download('omw-1.4')
```

```python
corpus = []
for i in range(len(data)):
    words = re.sub('[^a-zA-Z]', ' ', data["message"][i])
    words = words.lower()
    words = nltk.word_tokenize(words)
    # words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words("english"))]
    words = ' '.join(words)
    corpus.append(words)
```

```python
corpus
```

# tfidf vectorizing

```python
X = tfidfv.fit_transform(corpus).toarray()
```

```python
print(X.shape)
```

# assingning values to labeled data 

```python
Y = pd.get_dummies(data["lable"])
Y
```

```python
Y = Y.iloc[:, 1].values
```

# splitting data to training and testing set

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, random_state=0, test_size=0.2)
```

# SVM Model

```python
svc = SVC(kernel="linear")
model = svc.fit(xtrain, ytrain)
```

# prediction

```python
dtct = model.predict(xtest)
```

# checking accuracy

```python
cm = confusion_matrix(ytest, dtct)
acuuracy = accuracy_score(ytest, dtct)
print(acuuracy)
print(cm)
```

# NAIVE BAYES

```python
from sklearn.naive_bayes import MultinomialNB  # Naive bayes
# naive bayes algorithm for classification
MNN = MultinomialNB()
spam_detection_model = MNN.fit(xtrain, ytrain)
```

```python
# making predictions and checking accuracy score
detect = spam_detection_model.predict(xtest)
cm = confusion_matrix(ytest, detect)
acuuracy = accuracy_score(ytest, detect)
print(acuuracy)
print(cm)
```
